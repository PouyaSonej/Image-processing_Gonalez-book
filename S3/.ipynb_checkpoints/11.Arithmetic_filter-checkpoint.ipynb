```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img6 = cv2.imread("/content/Camera_man.png")
row,col,_ = img6.shape
s_vs_p = 0.5
amount = 0.1
out = np.copy(img6)
# Salt mode
num_salt = np.ceil(amount * img6.size * s_vs_p)
coords = [np.random.randint(0, i - 1, int(num_salt)) for i in img6.shape]
out[coords] = 1

# Pepper mode
num_pepper = np.ceil(amount* img6.size * (1. - s_vs_p))
coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in img6.shape]
out[coords] = 0
plt.imshow(out,cmap='gray')
```

```python
IMG3 = cv2.boxFilter(out, -1, (5, 5), normalize=True)  
plt.imshow(IMG3,cmap='gray')
```
