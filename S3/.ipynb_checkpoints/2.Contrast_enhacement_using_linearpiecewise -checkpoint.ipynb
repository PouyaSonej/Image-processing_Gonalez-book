```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Original Images
Img3 = cv2.imread('/content/Fig3.10(b).jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(Img3,cmap='gray')
```

```python
def linear_piecewise(Img3,x):
  img = Img3/255
  img = x * img
  img[img<(40/255)] = 2 * img[img<(40/255)]
  img[img>(100/255)] = 2 * img[img>(100/255)]
  return img

Img33 = linear_piecewise(Img3,0.44)
img333 = linear_piecewise(Img3,0.7)
img3333 = linear_piecewise(Img3,0.75)

plt.subplot(221)
plt.imshow(Img3,cmap="gray")
plt.title('Original')
plt.subplot(222)
plt.imshow(Img33,cmap='gray')
plt.title('ratio 0.5')
plt.subplot(223)
plt.imshow(img333,cmap='gray')
plt.title('ratio 0.7')
plt.subplot(224)
plt.imshow(img3333,cmap='gray')
plt.title('ratio 0.8')
```
