```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
img9 = cv2.imread("/content/Gradient.PNG")
img10 = cv2.cvtColor(img9, cv2.COLOR_BGR2GRAY)
out = cv2.normalize(img10.astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)


# # Get x-gradient in "sx"
# sx = ndimage.sobel(img10,axis=0,mode='constant')
# # Get y-gradient in "sy"
# sy = ndimage.sobel(img10,axis=1,mode='constant')
# # Get square root of sum of squares
# sobel=np.hypot(sx,sy)

# Define kernel for x differences
kx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
# Define kernel for y differences
ky = np.array([[1,2,1] ,[0,0,0], [-1,-2,-1]])
# Perform x convolution
x=ndimage.convolve(out,kx)
# Perform y convolution
y=ndimage.convolve(out,ky)
sobel=np.hypot(x,y)

# Hopefully see some edges
plt.subplot(121)
plt.imshow(img10,cmap="gray")
plt.axis("off")
plt.subplot(122)
plt.imshow(sobel,cmap="gray")
plt.axis("off")
plt.show()
```
