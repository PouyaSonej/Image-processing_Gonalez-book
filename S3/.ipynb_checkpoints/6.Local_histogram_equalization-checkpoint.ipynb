```python
from skimage.morphology import disk,ball,cube,square
from skimage.filters import rank
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
img2 = cv2.imread("/content/morrre.PNG")
# img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

footprint =  ball(3)
footprint1 =  cube(3)

img_eq = rank.equalize(img2, footprint=footprint)
img_eq1 = rank.equalize(img2, footprint=footprint1)

img_eq = cv2.cvtColor(img_eq, cv2.COLOR_BGR2GRAY)
img_eq1 = cv2.cvtColor(img_eq1, cv2.COLOR_BGR2GRAY)

plt.subplot(131)
plt.imshow(img2,cmap="gray")

plt.subplot(132)
plt.imshow(img_eq,cmap='gray')

plt.subplot(133)
plt.imshow(img_eq1,cmap='gray')

```
