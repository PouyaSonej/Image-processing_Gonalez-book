```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/3.33.PNG")

# Defining a kernel using numpy.
# kernel_7 = np.ones((35,35),np.float32)/1225
# kernel_15 = np.ones((15,15),np.float32)/225
# kernel_9 = np.ones((9,9),np.float32)/81
# kernel_5 = np.ones((5,5),np.float32)/25
# kernel_3 = np.ones((3,3),np.float32)/9

IMG = []
#border as constant equals to zero padding in matlab
IMG.append(cv2.boxFilter(img,-1,ksize=(3,3),borderType=cv2.BORDER_CONSTANT))
IMG.append(cv2.boxFilter(img,-1,ksize=(5,5),borderType=cv2.BORDER_CONSTANT))
IMG.append(cv2.boxFilter(img,-1,ksize=(7,7),borderType=cv2.BORDER_CONSTANT))
IMG.append(cv2.boxFilter(img,-1,ksize=(9,9),borderType=cv2.BORDER_CONSTANT))
IMG.append(cv2.boxFilter(img,-1,ksize=(15,15),borderType=cv2.BORDER_CONSTANT))
IMG.append(cv2.boxFilter(img,-1,ksize=(35,35),borderType=cv2.BORDER_CONSTANT))

#border as isolated equals as ignore borders in matlab
IMG.append(cv2.boxFilter(img,-1,ksize=(3,3),borderType=cv2.BORDER_ISOLATED ))
IMG.append(cv2.boxFilter(img,-1,ksize=(5,5),borderType=cv2.BORDER_ISOLATED ))
IMG.append(cv2.boxFilter(img,-1,ksize=(7,7),borderType=cv2.BORDER_ISOLATED ))
IMG.append(cv2.boxFilter(img,-1,ksize=(9,9),borderType=cv2.BORDER_ISOLATED ))
IMG.append(cv2.boxFilter(img,-1,ksize=(15,15),borderType=cv2.BORDER_ISOLATED ))
IMG.append(cv2.boxFilter(img,-1,ksize=(35,35),borderType=cv2.BORDER_ISOLATED ))

#border as replicate equals as replicate borders in matlab
IMG.append(cv2.boxFilter(img,-1,ksize=(3,3),borderType=cv2.BORDER_REPLICATE  ))
IMG.append(cv2.boxFilter(img,-1,ksize=(5,5),borderType=cv2.BORDER_REPLICATE  ))
IMG.append(cv2.boxFilter(img,-1,ksize=(7,7),borderType=cv2.BORDER_REPLICATE  ))
IMG.append(cv2.boxFilter(img,-1,ksize=(9,9),borderType=cv2.BORDER_REPLICATE  ))
IMG.append(cv2.boxFilter(img,-1,ksize=(15,15),borderType=cv2.BORDER_REPLICATE  ))
IMG.append(cv2.boxFilter(img,-1,ksize=(35,35),borderType=cv2.BORDER_REPLICATE  ))
plt.figure(figsize=(10,10))
for i in range(18):
    plt.subplot(3, 6, i + 1)
    # Convert [0 - 255] to [0.0 - 1.0]
    # out[i] = cv2.normalize(img22[i].astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
    plt.imshow(np.array(IMG[i]),cmap='gray')
    plt.axis("off")
plt.show()

```
