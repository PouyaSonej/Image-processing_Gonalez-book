```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python


img3 = cv2.imread("/content/Capture.PNG")
img4 = cv2.imread("/content/Picture1.png")

clahe = cv2.createCLAHE(clipLimit =5.0, tileGridSize=(8,8))
cl_img1 = clahe.apply(img3[:,:,1])
cl_img2 = clahe.apply(img4[:,:,1])

plt.figure(figsize=(10,5))
plt.subplot(241)
plt.imshow(img3,cmap="gray")
plt.axis("off")

plt.subplot(242)
plt.imshow(cl_img1,cmap="gray")
plt.axis("off")

plt.subplot(243)
plt.imshow(img4,cmap="gray")
plt.axis("off")

plt.subplot(244)
plt.imshow(cl_img2,cmap="gray")
plt.axis("off")
plt.subplot(245)
plt.hist(img3.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

plt.subplot(246)
plt.hist(cl_img1.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

plt.subplot(247)
plt.hist(img4.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")
plt.subplot(248)
plt.hist(cl_img2.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

clear_output(wait=True)
```
