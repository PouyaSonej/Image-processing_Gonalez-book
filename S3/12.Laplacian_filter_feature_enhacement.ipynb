```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python

img8 = cv2.imread("/content/MOON.PNG")

kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=img8, ddepth=-1, kernel=kernel)

kernel1 = np.array([[-1, -1, -1],
                   [-1, 9,-1],
                   [-1, -1, -1]])
image_sharp1 = cv2.filter2D(src=img8, ddepth=-1, kernel=kernel1)

plt.subplot(131)
plt.imshow(img8)

plt.subplot(132)
plt.imshow(image_sharp)

plt.subplot(133)
plt.imshow(image_sharp1)
```
