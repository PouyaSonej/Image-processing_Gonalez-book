```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Original Images
img5=[]
img5.append(cv2.imread('/content/Fig3.15(a)3.jpg', cv2.IMREAD_GRAYSCALE))
img5.append(cv2.imread('/content/Fig3.20(a).jpg', cv2.IMREAD_GRAYSCALE))
    
    
k=0
f,ax2=plt.subplots(2,2,figsize=(20,20))
for i in range(2) :
    ax2[i,0].hist(img5[k].ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
    ax2[i,1].imshow(img5[k],cmap="gray")
    k+=1
```

```python
image = img5[0]
hist,bins = np.histogram(image.flatten(),256,[0,256])
image = cv2.equalizeHist(image[:,:])
plt.subplot(121)
plt.hist(image.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.subplot(122)
plt.imshow(image,cmap="gray")
```
