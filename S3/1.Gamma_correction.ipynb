```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Original Images
Img1 = cv2.imread('/content/Fig3.08(a).jpg', cv2.IMREAD_GRAYSCALE)
Img2 = cv2.imread('/content/Fig3.09(a).jpg', cv2.IMREAD_GRAYSCALE)
    
    
plt.subplot(121)
plt.imshow(Img1,cmap="gray")
plt.subplot(122)
plt.imshow(Img2,cmap="gray")
```

```python
#First image

#gamma = 9.0
gamma_corrected1 = np.array(255*(Img1 / 255) ** 9, dtype = 'uint8')
#gamma = 1.5
gamma_corrected2 = np.array(255*(Img1 / 255) ** 1.5, dtype = 'uint8')
#gamma = 0.2
gamma_corrected3 = np.array(255*(Img1 / 255) ** 0.5, dtype = 'uint8')

plt.subplot(221)
plt.imshow(Img1,cmap="gray")
plt.title('Original')
plt.subplot(222)
plt.imshow(gamma_corrected1,cmap='gray')
plt.title('gamma = 9.0')
plt.subplot(223)
plt.imshow(gamma_corrected2,cmap='gray')
plt.title('gamma = 1.5')
plt.subplot(224)
plt.imshow(gamma_corrected3,cmap='gray')
plt.title('gamma = 0.2')
```

```python
#Second image

#gamma = 9.0
gamma_corrected1 = np.array(255*(Img2 / 255) ** 9, dtype = 'uint8')
#gamma = 1.5
gamma_corrected2 = np.array(255*(Img2 / 255) ** 1.5, dtype = 'uint8')
#gamma = 0.2
gamma_corrected3 = np.array(255*(Img2 / 255) ** 0.5, dtype = 'uint8')

plt.subplot(221)
plt.imshow(Img2,cmap="gray")
plt.title('Original')
plt.subplot(222)
plt.imshow(gamma_corrected1,cmap='gray')
plt.title('gamma = 9.0')
plt.subplot(223)
plt.imshow(gamma_corrected2,cmap='gray')
plt.title('gamma = 1.5')
plt.subplot(224)
plt.imshow(gamma_corrected3,cmap='gray')
plt.title('gamma = 0.2')
```
