```python
import skimage
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#using higher samples(resolution like (1024,1024)) improves OP

#moire
image3 = np.ones((130,130))

for i in range(0,130):
  for j in range(0,130,2):
    image3[i,j] = 0

# use -15 instead of -45 for more acccurate OP
rotated=skimage.transform.rotate(image3,-45,resize=False)
cropped1 = rotated[20:110, 20:110]
cropped3 = cropped1 + image3[20:110, 20:110]
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(image3,cmap='gray')
plt.title("pattern")
plt.axis('off')
plt.subplot(222)
plt.imshow(rotated,cmap='gray')
plt.title("rotated 45 degrees")
plt.axis('off')
plt.subplot(223)
plt.imshow(cropped3,cmap='gray')
plt.title("cropped")
plt.axis('off')
plt.subplot(224)
plt.imshow(cropped1,cmap='gray')
plt.title("combined")
plt.axis('off')
```
