```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
```

```python
img2 = plt.imread("/content/barbara.jpg")

scale_percent = 50 # percent of original size
width = int(img2.shape[1] * scale_percent / 100)
height = int(img2.shape[0] * scale_percent / 100)
dim = (width, height)

image1 = cv2.resize(img2,dsize=dim,interpolation = cv2.INTER_NEAREST )

plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img2,cmap='gray')
plt.title(img2.shape)
# plt.axis('off')
plt.subplot(122)
plt.imshow(image1,cmap='gray')
plt.title(image1.shape)
# plt.axis('off')
plt.ylabel("50% smaller scale")
```

```python
#gaussian filter using diffret sigma
# .1
image2_1 = gaussian_filter(img2, sigma=.1)
# 1
image2_2 = gaussian_filter(img2, sigma=1)
# 2
image2_3 = gaussian_filter(img2, sigma=2)
plt.figure(figsize=(8,8))
plt.subplot(221)
plt.imshow(img2,cmap='gray')
plt.axis('off')
plt.subplot(222)
plt.imshow(image2_1,cmap='gray')
plt.axis('off')
plt.subplot(223)
plt.imshow(image2_2,cmap='gray')
plt.axis('off')
plt.subplot(224)
plt.imshow(image2_3,cmap='gray')
plt.axis('off')
```
