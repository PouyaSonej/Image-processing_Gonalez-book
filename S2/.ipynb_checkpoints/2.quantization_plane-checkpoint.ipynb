```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Original image
img2 = cv2.imread("/content/Fig2.21(a).jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img2,cmap='gray')
img2.shape
```

```python
#Output
img22 = []
for i in range(0,8):
  n = cv2.normalize(img2, None, 0, pow(2,i), cv2.NORM_MINMAX)
  img22.append(n)
for i in range(8):
    plt.subplot(2, 4, i + 1)
    # Convert [0 - 255] to [0.0 - 1.0]
    # out[i] = cv2.normalize(img22[i].astype('float'), None, 0.0, 1.0, cv2.NORM_MINMAX)
    plt.imshow(np.array(img22[i]),cmap='gray')
    plt.title(f"quan plane {i}")
plt.show()
```
