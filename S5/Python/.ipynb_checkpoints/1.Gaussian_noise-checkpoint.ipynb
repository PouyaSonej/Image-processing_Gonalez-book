```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage  # Used skimage as noise functions
from IPython.display import clear_output
```

```python
#Original image
img1 = cv2.imread('/content/Fig0503 (original_pattern).tif',cv2.IMREAD_GRAYSCALE)
plt.imshow(img1,cmap='gray')
plt.axis("off")

img1.shape
```

```python
#Gaussian noise
mean = 1
var = 60
sigma = var ** 0.5
gaussian = np.random.normal(mean, sigma, (256, 256)) #  np.zeros((224, 224), np.float32)
GN1 = img1 + gaussian

#Exponential noise
noise_std = 0.2
image = img1.astype(np.float64)
noise = np.random.rayleigh(noise_std, img1.shape)
RN1 = cv2.addWeighted(image, 1, noise, 70, 0.0).astype(np.uint8)

#Uniform noise
uni_noise=np.zeros((256,256),dtype=np.uint8)
cv2.randu(uni_noise,0,255)
uni_noise=(uni_noise*.09).astype(np.uint8)
UN1 = cv2.add(img1,uni_noise)
# s = np.random.gamma(256,256)
# GN2 = img1 + s
```

```python
plt.figure(figsize=(10,5))
plt.subplot(241)
plt.imshow(img1,cmap="gray")
plt.axis("off")

plt.subplot(242)
plt.imshow(GN1,cmap="gray")
plt.axis("off")

plt.subplot(243)
plt.imshow(RN1,cmap="gray")
plt.axis("off")

plt.subplot(244)
plt.imshow(UN1,cmap="gray")
plt.axis("off")
plt.subplot(245)
plt.hist(img1.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

plt.subplot(246)
plt.hist(GN1.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

plt.subplot(247)
plt.hist(RN1.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")
plt.subplot(248)
plt.hist(UN1.ravel(), bins=256, range=(0.0, 255.0), fc='k', ec='k')
plt.axis("off")

clear_output(wait=True)
```
