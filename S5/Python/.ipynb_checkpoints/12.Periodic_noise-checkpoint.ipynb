```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage  # Used skimage as noise functions
from IPython.display import clear_output
```

```python
#Original image
img7 = cv2.imread('/content/cameraman.tif',cv2.IMREAD_GRAYSCALE)
plt.imshow(img7,cmap='gray')
plt.axis("off")

img7.shape
```

```python
img = cv2.normalize(img7.astype("float"), None, 0.0, 1.0, cv2.NORM_MINMAX)
pi = np.pi
# Create periodic noise

# img shape = 256, 256
[m, n] = img.shape
[a, b] = np.meshgrid(np.arange(0, n), np.arange(0, m))
p = np.sin(a/4) + np.cos(b/4) + 1
# Adding made noise to the image
imgn = cv2.normalize((img + p) /2, None, 0, 255, cv2.NORM_MINMAX)
plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title("Image without noise")
plt.subplot(122)
plt.imshow(imgn, cmap='gray')
plt.title("Image with noise")
plt.show()
```
