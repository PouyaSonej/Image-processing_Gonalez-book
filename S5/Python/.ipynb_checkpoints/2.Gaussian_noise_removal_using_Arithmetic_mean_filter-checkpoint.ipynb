```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import skimage  # Used skimage as noise functions
from IPython.display import clear_output
```

```python
#Original image
img2 = cv2.imread('/content/Fig0507(a)(ckt-board-orig).tif',cv2.IMREAD_GRAYSCALE)
plt.imshow(img2,cmap='gray')
plt.axis("off")

img2.shape
```

```python
#Gaussian noise
mean = 0
var = 0.05
sigma = var ** 0.5
gaussian = np.random.normal(mean, sigma, (448,464)) #  np.zeros((224, 224), np.float32)
GN2 = img2 + gaussian

GN3 = skimage.util.random_noise(img2, 'gaussian', mean=0, var=0.05)

image = GN2.astype(float)
rows, cols = image.shape[:2]
ksize = 5
padsize = int((ksize-1)/2)
pad_img = cv2.copyMakeBorder(image, *[padsize]*4, cv2.BORDER_DEFAULT)
geomean1 = np.zeros_like(image)
for r in range(rows):
    for c in range(cols):
        geomean1[r, c] = np.prod(pad_img[r:r+ksize, c:c+ksize])**(1/(ksize**2))
geomean1 = np.uint8(geomean1)

# Arithmetic mean filter
arithmetic1 = cv2.boxFilter(image, -1, (5, 5), normalize=True)  


plt.figure(figsize=(10,5))
plt.subplot(141)
plt.imshow(img2,cmap="gray")
plt.title("org")
plt.axis("off")

plt.subplot(142)
plt.imshow(GN3,cmap="gray")
plt.title("noisy")
plt.axis("off")

plt.subplot(143)
plt.imshow(geomean1,cmap="gray")
plt.title("geo")
plt.axis("off")

plt.subplot(144)
plt.imshow(arithmetic1,cmap="gray")
plt.title("Arith")
plt.axis("off")
```
